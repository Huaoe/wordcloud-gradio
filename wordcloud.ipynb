```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, ImageColorGenerator
import re
import numpy as np
from PIL import Image
```

# Import des données

```python
#importer les données textuelles
df = pd.read_excel("IMDB-Dataset.xlsx")

#Variable texte
vt = " ".join(review for review in df.review)
```

```python
#Création de mon nuage de mots avec l'argument texte dans la méthode .generate()
word_cloud = WordCloud(collocations=False, background_color='white', width=500, height=500).generate(vt)
```

```python
#Exporter au format PNG 
word_cloud.to_file("testwordcloud.png")
```

```python
#afficher le wordcloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Import du texte
dataset = open("text.txt", "r").read()
```

```python
#Affecter une image pour le masque
background_image = np.array(Image.open("morrade-decontoure.jpg"))
plt.imshow(background_image)
```

```python
#Création du nuage de mots
word_cloud2 = WordCloud(background_color='white', mask=background_image, width=800, height=1250).generate(dataset)

img_colors = ImageColorGenerator(background_image)
word_cloud2.recolor(color_func= img_colors)
```

```python
# sauvegarder l'Image
word_cloud2.to_file("boxing.png")
#afficher le wordcloud
plt.imshow(word_cloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
